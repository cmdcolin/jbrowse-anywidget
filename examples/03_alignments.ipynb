## Assembly and alignments

The CRAM's `.crai` index and its reference sequence are resolved automatically from the `uri`, so the adapter is just the URL.

```python
from jbrowse_anywidget import LinearGenomeView, make_assembly

grch38 = make_assembly(
    "GRCh38",
    "https://s3.amazonaws.com/jbrowse.org/genomes/GRCh38/fasta/GRCh38.fa.gz",
    aliases=["hg38"],
)

cram = (
    "https://s3.amazonaws.com/jbrowse.org/genomes/GRCh38/alignments/NA12878/"
    "NA12878.alt_bwamem_GRCh38DH.20150826.CEU.exome.cram"
)

view = LinearGenomeView(
    assembly=grch38, location="1:100,987,200..100,987,450"
)
view.add_track(
    {
        "type": "AlignmentsTrack",
        "trackId": "na12878-exome",
        "name": "NA12878 exome",
        "assemblyNames": ["GRCh38"],
        "adapter": {"type": "CramAdapter", "uri": cram},
    }
)
view
```

## Color reads, show soft-clips

A track config can carry a `displays` entry to preset the display — here color by pair orientation to surface structural signal, and reveal soft-clipped bases.

```python
view.add_track(
    {
        "type": "AlignmentsTrack",
        "trackId": "na12878-colored",
        "name": "NA12878 (pair orientation)",
        "assemblyNames": ["GRCh38"],
        "adapter": {"type": "CramAdapter", "uri": cram},
        "displays": [
            {
                "type": "LinearAlignmentsDisplay",
                "displayId": "na12878-colored-display",
                "colorBy": {"type": "pairOrientation"},
                "showSoftClipping": True,
            }
        ],
    }
)
```
